```python
import pandas as pd
import numpy as np
```

```python
# 텍스트 파일 경로
file_path = '/content/drive/MyDrive/2023_서동석교수님_랩인턴/Raman/Kim_S1/0110_1_3 con.txt'

raman = pd.read_csv(file_path, header=None, sep = '\t')

# 내용 출력
raman.columns = ["X", "Y"]
raman
```

```python
import matplotlib.pyplot as plt

# Define wavelength ranges
range1 = (2400, 2700)
range2 = (1500, 1800)

# Filter data within specified wavelength ranges
df_range1 = raman[(raman['X'] >= range1[0]) & (raman['X'] <= range1[1])]
df_range2 = raman[(raman['X'] >= range2[0]) & (raman['X'] <= range2[1])]

plt.figure()
plt.plot(raman['X'], raman['Y'], color=(252/255, 103/255, 54/255))

# Setting font size for x and y axis labels
plt.xlabel('Raman Shift (1/cm)', fontsize=13)
plt.ylabel('Intensity', fontsize=13)
plt.title('Raman Spectrum')
plt.grid(True)

# Setting font size for ticks (numbers) and adjusting x tick spacing and start point
plt.xticks(ticks=range(1200, int(max(raman['X']))+1, 300), fontsize=15)
plt.yticks(fontsize=15)

# Find highest peak in range 1
peak1_range1_index = df_range1['Y'].idxmax()
peak1_range1_x = df_range1.loc[peak1_range1_index, 'X']
peak1_range1_y = df_range1.loc[peak1_range1_index, 'Y']

# Find highest peak in range 2
peak1_range2_index = df_range2['Y'].idxmax()
peak1_range2_x = df_range2.loc[peak1_range2_index, 'X']
peak1_range2_y = df_range2.loc[peak1_range2_index, 'Y']

# Plot peaks on the graph
plt.scatter(peak1_range1_x, peak1_range1_y, color='green', label='2D')
plt.scatter(peak1_range2_x, peak1_range2_y, color='blue', label='G')

# Annotate the peaks with their intensity values
plt.annotate(f'2D: {peak1_range1_y:.2f}', (peak1_range1_x, peak1_range1_y), textcoords="offset points", xytext=(-10,-20), ha='right', fontsize=12)
plt.annotate(f'G: {peak1_range2_y:.2f}', (peak1_range2_x, peak1_range2_y), textcoords="offset points", xytext=(10,-15), ha='left', fontsize=12)

# Show legend
plt.legend()
# X 축 범위 설정
plt.xlim(1200, 3200)

plt.show()
```
