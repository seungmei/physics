```python
import pandas as pd
import numpy as np
```

```python
# 텍스트 파일 경로
file_path = '/content/drive/MyDrive/2023_서동석교수님_랩인턴/Raman/Kim_S1/0110_1_3 con.txt'

raman = pd.read_csv(file_path, header=None, sep = '\t')

# 내용 출력
raman.columns = ["X", "Y"]
raman
```

```python
import matplotlib.pyplot as plt

# 데이터 플로팅
plt.figure(figsize=(10, 6))
plt.plot(raman['X'], raman['Y'], marker='o', linestyle='-')
plt.title('Raman Spectrum')
plt.xlabel('Raman Shift (1/cm)')
plt.ylabel('Intensity')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터 플로팅
plt.figure(figsize=(10, 6))
plt.plot(raman['X'], raman['Y'], marker='o', linestyle='-')
plt.title('Raman Spectrum')
plt.xlabel('Raman Shift (1/cm)')
plt.ylabel('Intensity')
plt.grid(True)

# X 축 범위 설정
plt.xlim(1200, 3200)

# X가 100 증가할 때마다 해당 위치의 Intensity 값을 표시
for i, txt in enumerate(raman['Y']):
    if i % 100 == 0:
        plt.annotate(f'{txt:.2f}', (raman['X'][i], txt), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```
