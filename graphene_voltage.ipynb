```python
import pandas as pd
import numpy as np
```

```python
file = '/content/drive/MyDrive/2023_서동석교수님_랩인턴/측정/4. SMH_Vg_0.01 cd'

# 데이터프레임 생성
df = pd.DataFrame(np.loadtxt(file))
df
```

```python
# 열 이름 변경
df.columns = ['Time', 'Vg', 'Ileak', 'Vsd', 'Isd', 'R']

# 변경된 데이터프레임 출력
df
```

```python
import matplotlib.pyplot as plt

# Plotting Isd with lines and markers
plt.plot(df['Vg'], df['Isd'], label='Isd', marker='o', linestyle='-', color='blue', linewidth=0.8)

# Adding labels and title
plt.xlabel('Vg')
plt.ylabel('Isd')
plt.title('Isd vs Vg with Lines and Markers')
plt.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd

# Assuming your original DataFrame is named 'df'
# You can replace 'df' with the actual name of your DataFrame if it's different

# Create a list to store the split DataFrames
split_dfs = []

# Set the initial index and the step size
index = 0
step_size = 21

# Iterate through the rows and split the DataFrame
while index < len(df):
    split_dfs.append(df.iloc[index:index+step_size])
    index += step_size
    step_size += 20  # Increase the step size for the next split DataFrame

# Create new DataFrame names (df1, df2, df3, ...)
df_names = [f'df{i+1}' for i in range(len(split_dfs))]

# Assign the split DataFrames to the new names
for name, split_df in zip(df_names, split_dfs):
    globals()[name] = split_df
```
