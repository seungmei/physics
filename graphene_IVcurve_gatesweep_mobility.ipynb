## gate sweep 구간별로 split 후에 그래프를 plot 해보자.
### gate sweep 구간별로 split 해주기 위해 규칙을 알아보니 gate sweep 구간이 21개의 행으로 시작하여 다음 구간으로 넘어갈때마다 20개씩 데이터의 행이 증가하는 규칙이 있음을 알 수 있다.


```python
import pandas as pd

# Assuming your original DataFrame is named 'df'
# You can replace 'df' with the actual name of your DataFrame if it's different

# Create a list to store the split DataFrames
split_dfs = []

# Set the initial index and the step size
index = 0
step_size = 21

# Iterate through the rows and split the DataFrame
while index < len(GW):
    split_dfs.append(GW.iloc[index:index+step_size])
    index += step_size
    step_size += 20  # Increase the step size for the next split DataFrame

# Create new DataFrame names (df1, df2, df3, ...)
df_names = [f'df{i+1}' for i in range(len(split_dfs))]

# Assign the split DataFrames to the new names
for name, split_df in zip(df_names, split_dfs):
    globals()[name] = split_df
```

# mobility를 계산해보자

## Drude model
 J = n * e * (mobility) * E
 선형의 관계일 경우에는 이렇게 바로 구해도 가능
##
n = Cg * Vg / e

(mobility) = (1 / Cg) * (d Isd/ d Vg) * (1 / Vsd) * ((Length) / (Width))

(Length) / (Width) ~ 0.8

### dataset = df1

```python
# Calculate dI/dVg
df1['dI/dVg'] = np.gradient(df1['Isd']*1e6, df1['Vg'])
```

```python
## Apply Savitzky-Golay filter
from scipy.signal import savgol_filter

window_lenght = 5
polyorder = 2
df1['dI/dVg_smoothing'] = savgol_filter(df1['dI/dVg'], window_lenght, polyorder)

# Plotting original dI/dVg
plt.figure()
plt.plot(df1['Vg'], df1['dI/dVg'], label='dI/dVg Original')
plt.xlabel('Gate Voltage (Vg)')
plt.ylabel('dI/dVg')
plt.title('Original vs. Smoothed dI/dVg')
plt.legend()
```

```python
import matplotlib.pyplot as plt

# Constants
Cg = 1.15 * 1e-8  # F/cm

# Normalize the RGB values (252, 103, 54) to [0, 1]
color_normalized = (100/255, 103/255, 54/255)

# Ensure you're using 'df1['Vsd']' if 'Vsd' values are relevant to the sliced portion 'df1'
# Calculate carrier mobility
# Adjusting the formula based on the assumption above
carrier_mobility = 0.8/ Cg * df1['dI/dVg']/df1['Vsd']
# Plotting
plt.figure()
plt.plot(df1['Vg'], carrier_mobility, label='Carrier Mobility', color='red')
plt.xlabel('Gate Voltage (Vg)')
plt.ylabel('Carrier Mobility (cm$^2$/Vs)')
plt.title('Carrier Mobility vs. Gate Voltage')
plt.legend()
plt.show()
```
